```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
from sklearn.compose import ColumnTransformer
```

```python
data.rename(columns={'WindGustDir':'Wind_Dir'}, inplace=True)
```

```python
data.dropna(subset=['Wind_Dir'],inplace=True)
data.dropna(subset=['RainToday'],inplace=True)
data.dropna(subset=['RainTomorrow'],inplace=True)
```

```python
a = ['Wind_Dir']
b = ['RainToday', 'RainTomorrow']
```

```python
categories = [['No', 'Yes'], ['No','Yes']]
```

```python
processor = ColumnTransformer([('Label', oe, a), ('Ordinal Encoder', OrdinalEncoder(categories=categories),b)])
```

```python
import pickle
file = open('Rain.pkl', 'wb')
pickle.dump(processor,file)
file.close()
```
